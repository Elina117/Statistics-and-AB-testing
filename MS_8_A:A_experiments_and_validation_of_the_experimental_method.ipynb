```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

iterations = 1000
pvalues = []
for _ in range(iterations):

    group1 = np.random.binomial(1000, 0.2, size=1000)
    group2 = np.random.binomial(1000, 0.2, size=1000)
    t_stat, pvalue = stats.ttest_ind(group1, group2)
    pvalues.append(pvalue)


plt.hist(pvalues, bins=50, edgecolor='black', alpha=0.7)
plt.xlabel("p-value")
plt.ylabel("")
plt.title(" p-value")
plt.show()
```

```python
sum_bigger_alpha = 0
sum_smaller_alpha = 0

for i in range(len(pvalues)):
    if pvalues[i] > 0.05:
        sum_bigger_alpha+=1

    else:
        sum_smaller_alpha+=1

print(f'   0.05: {sum_smaller_alpha}')
print(f'   0.05: {sum_bigger_alpha}')
```

```python
print(f'  : {sum_smaller_alpha/len(pvalues)}')
```

```python

alpha = 0.05
betta = 0.9
D = 10
effect = 0.5

for df in [10, 100, 1000, 10000]:
    t_alpha = stats.t.ppf(1 - alpha/2, df=df)
    t_betta = stats.t.ppf(betta, df=df)

    n = 2 * (t_alpha +t_betta)**2 * D / effect**2
    print(f'    df={df}: {n}')
```
